```python
import pandas as pd
from sqlalchemy import create_engine
import plotly.express as px
```

```python
#conexao
usuario = 'root'
senha = '****************'
host = 'localhost'
banco = 'apex_telemetria'

#string de conexao
engine = create_engine(f'mysql+mysqlconnector://{usuario}:{senha}@{host}/{banco}')
```

```python
#query para trazer os dados do banco para o python
query = "SELECT e.nome,v.* FROM consulta_manutencao AS v INNER JOIN equipes AS e ON v.id_veiculo = e.id_equipe ORDER BY id_veiculo ASC;"
```

```python
df_manutencao = pd.read_sql(query, engine)
colunas_valores = ["custo_estimado_preventiva","custo_real_preventiva","custo_estimado_corretiva","custo_real_corretiva","custo_estimado_total","custo_real_total"]
df_manutencao[colunas_valores] = df_manutencao[colunas_valores].fillna(0)
df_manutencao
```

```python
fig = px.bar(df_manutencao, x="id_veiculo", y=["custo_real_corretiva", "custo_real_preventiva"], barmode="group")
fig.update_xaxes(type='category')
fig.show()
```
